```python
# create a new dataframe
rows = []
```

```python
from tqdm import tqdm
```

```python
# for var in common_columns:
for var in tqdm(common_columns):
    print(var)

    svi_2014_temp = svi_2014[svi_2014[var] != -999]
    svi_2016_temp = svi_2016[svi_2016[var] != -999]
    svi_2018_temp = svi_2018[svi_2018[var] != -999]
    svi_2020_temp = svi_2020[svi_2020[var] != -999]
    svi_2022_temp = svi_2022[svi_2022[var] != -999]

    for STCNTY in common_STCNTY:
        # print(STCNTY)
        mean_2014 = svi_2014_temp[svi_2014_temp['STCNTY'] == STCNTY][var].mean()
        # add this to the dataframe
        rows.append({'STCNTY': STCNTY, 'variable': var, 'year': 2014, 'mean': mean_2014})

        mean_2016 = svi_2016_temp[svi_2016_temp['STCNTY'] == STCNTY][var].mean()
        # add this to the dataframe
        rows.append({'STCNTY': STCNTY, 'variable': var, 'year': 2016, 'mean': mean_2016})

        mean_2018 = svi_2018_temp[svi_2018_temp['STCNTY'] == STCNTY][var].mean()
        # add this to the dataframe
        rows.append({'STCNTY': STCNTY, 'variable': var, 'year': 2018, 'mean': mean_2018})

        mean_2020 = svi_2020_temp[svi_2020_temp['STCNTY'] == STCNTY][var].mean()
        # add this to the dataframe
        rows.append({'STCNTY': STCNTY, 'variable': var, 'year': 2020, 'mean': mean_2020})

        mean_2022 = svi_2022_temp[svi_2022_temp['STCNTY'] == STCNTY][var].mean()
        # add this to the dataframe
        rows.append({'STCNTY': STCNTY, 'variable': var, 'year': 2022, 'mean': mean_2022})
```

```python
# Convert the list of dictionaries to a DataFrame
svi_info_county_yearly = pd.DataFrame(rows)
```

```python
# census level analysis

# for var in common_columns:
for var in tqdm(common_columns):
    print(var)

    svi_2014_temp = svi_2014[svi_2014[var] != -999]
    svi_2016_temp = svi_2016[svi_2016[var] != -999]
    svi_2018_temp = svi_2018[svi_2018[var] != -999]
    svi_2020_temp = svi_2020[svi_2020[var] != -999]
    svi_2022_temp = svi_2022[svi_2022[var] != -999]


    # cerate a new dataframe using this sci data where we have column variable, year, mean, FIPS, variable_value

    rows = []

    for r in svi_2014_temp.iterrows():
        rows.append({'FIPS': r[1]['FIPS'], 'variable': var, 'year': 2014, 'variable_value': r[1][var]})
    for r in svi_2016_temp.iterrows():
        rows.append({'FIPS': r[1]['FIPS'], 'variable': var, 'year': 2016, 'variable_value': r[1][var]})
    for r in svi_2018_temp.iterrows():
        rows.append({'FIPS': r[1]['FIPS'], 'variable': var, 'year': 2018, 'variable_value': r[1][var]})
    for r in svi_2020_temp.iterrows():
        rows.append({'FIPS': r[1]['FIPS'], 'variable': var, 'year': 2020, 'variable_value': r[1][var]})
    for r in svi_2022_temp.iterrows():
        rows.append({'FIPS': r[1]['FIPS'], 'variable': var, 'year': 2022, 'variable_value': r[1][var]})
    
    # Convert the list of dictionaries to a DataFrame
    svi_info_census_yearly = pd.DataFrame(rows)

    #save the dataframe to a csv file
    svi_info_census_yearly.to_csv(f'/home/h6x/git_projects/svi_data_analysis/data/processed_data/analysis_svi/{var}_census_2014_2022.csv', index=False)
```
