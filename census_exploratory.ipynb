```python
# Load the libraries
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
svi_age = pd.read_csv('/home/h6x/git_projects/ornl-svi-mortality-data-analysis/data/processed_data/analysis_svi/EP_AGE17_census_2014_2022.csv',dtype={'FIPS':str})
```

```python
svi_age.head()
```

```python
unique_fips = svi_age['FIPS'].unique()
```

```python
len(unique_fips)
```

```python
# plot the variable_value for each year and each FIPS
i=0

for fips in unique_fips:

    svi_age_fips = svi_age[svi_age['FIPS'] == fips]

    plt.plot(svi_age_fips['year'], svi_age_fips['variable_value'], label=fips)

    i+=1
    if i==15:
        break

# plt.legend()
plt.show()
```
